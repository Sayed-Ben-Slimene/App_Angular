```python
data = pd.read_csv('professional_profiles (6).csv', sep=';')
```

```python
pd.set_option('display.max_rows', None)

# Define a custom style function for the DataFrame
def style_df(df):
    return df.head(50).style.set_table_styles([{
        'selector': 'thead th',
        'props': [('background-color', '#4CAF50'), ('color', 'white')]
    }])

# Display the styled DataFrame
style_df(data)
```

```python
# Handling missing values
data.dropna(inplace=True)

# Removing duplicates
data.drop_duplicates(inplace=True)
```
